```python
def calcular_modelo (m,x,b):
    #modelo lineal
    ycalculada = (m*x)+(b+1)
    return ycalculada
```

```python
def gradiente_descendente (m_,b_,learning_rate,x,y,iteraciones):
    #gradiente para minimizar el error cuadratico
    #n = iteraciones #cantidad de datos
    #errores derivadas de la funcion de error con respecto a los parametros m y b
    #dm = -(2-n)*(np.sum(x*(y-(m_*x*b_)))) # -(2-n)*
    #db = -(2-n)*(np.sum(y-(m_*x*b_)))  # -(2-n)*
    
    dm =  (np.sum(x*(y-m_)))*((2/iteraciones)*-1)
    db =  (np.sum((y-m_)))  *((2/iteraciones)*-1) 
    
    
    #para graficar linea de errores respecto a las iteraciones
    m = m_ -learning_rate*dm
    b = b_ -learning_rate*db
     #m = np.float128(m)
     #b = np.float128(b)
    return m,b
```

```python
def calcular_error (y,yhat):
    #calcula el error del dato original y el dato generado por el modelo
    #n = y.shape[0]
    error = (np.sum((y-yhat)**2))/2
    #error = np.average((yhat-y)**2)*0.5
    return error
```

# scikit-learn

```python
import numpy as np
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() 
from sklearn.model_selection import train_test_split 
from sklearn import metrics
```

```python

x_train = np.array(datos_de_entrenamiento['OverallQual'].values)
y_train = np.array(datos_de_entrenamiento['SalesPrice'].values)
x_test  = np.array(datos_de_validacion_y_pruebas['OverallQual'].values)

x_train= x_train.reshape(-1, 1)
y_train= y_train.reshape(-1, 1)
x_test = y_train.reshape(-1, 1)

regressor.fit(x_train, y_train) #training the algorithm
print(regressor.intercept_)

#For retrieving the slope:
print(regressor.coef_)
y_pred = regressor.predict(x_test)
```

```python
df = pd.DataFrame({'Actual': y_train.flatten(), 'Predicted': y_pred.flatten()})
df
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_train, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_train, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_train, y_pred)))
sk1 = np.sqrt(metrics.mean_squared_error(y_train, y_pred))
```
